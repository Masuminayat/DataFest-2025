```python
#Page 2
#Master Slicer for year range selection

import ipywidgets as widgets
from IPython.display import display
import pandas as pd

# --- 1. Create the Master List of Years ---
# We iterate through all loaded dataframes to find every unique year string available.
all_year_ranges = set()
print("Scanning files for 'Year' data...")
for name, df in dataframes.items():
    if 'Year' in df.columns:
        # Convert to string, strip whitespace, drop NaNs, get unique values
        years = df['Year'].dropna().astype(str).str.strip().unique().tolist()
        all_year_ranges.update(years)
        print(f" - Found {len(years)} unique years in {name}.xlsx")

# Sort the years alphabetically for the dropdown list
sorted_years_options = sorted(list(all_year_ranges))

# --- 2. Create the Multi-Select Widget ---
# This widget replicates the Power BI slicer shown in your screenshots
# It allows selecting multiple options (e.g., holding Ctrl/Shift or just clicking on mobile).
year_slicer_widget = widgets.SelectMultiple(
    options=sorted_years_options,
    # Default selection: let's pick the last 2 available years as a starting point
    value=sorted_years_options[-2:] if len(sorted_years_options) >= 2 else sorted_years_options,
    rows=10, # How many rows to show at once in the box
    description='Year Range:',
    disabled=False,
    layout=widgets.Layout(width='300px')
)

print("\n--- Master Year Slicer Created ---")
print("Select year ranges below. These selections will be passed as variables to your future KPIs.")
display(year_slicer_widget)
```

```python
# helper function for year selection
# --- STANDARD HELPER FUNCTION ---
def filter_by_selected_years(df, selected_years_variable, year_col_name='Year'):
    """
    Global helper function to filter any dataframe based on the
    variable passed from the UI slicer.
    """
    # 1. Create a copy to avoid SettingWithCopy warnings
    df_filtered = df.copy()

    # 2. Ensure the dataframe's year column is clean string format for comparison
    df_filtered['temp_year_str'] = df_filtered[year_col_name].astype(str).str.strip()

    # 3. Filter: Keep rows where the year matches one of the selected variables
    # The 'selected_years_variable' is a tuple of strings, e.g., ('2018-2019', '2019-2020')
    df_final = df_filtered[df_filtered['temp_year_str'].isin(selected_years_variable)]

    # Clean up temp column
    df_final = df_final.drop(columns=['temp_year_str'])

    return df_final


# --- EXAMPLE: FUTURE KPI STRUCTURE ---
# This is a template for how you will write your future KPIs.
# Notice it accepts 'selected_years_tuple' as an argument.

def calculate_example_kpi(df, selected_years_tuple):
    # 1. Apply the standard filter using the passed variable
    filtered_df = filter_by_selected_years(df, selected_years_tuple)

    # 2. Perform calculation (e.g., count rows for this example)
    kpi_result = len(filtered_df)

    print(f"--- Calculating KPI for selected years: {selected_years_tuple} ---")
    print(f"Dataframe rows after filtering: {kpi_result}")
    # return kpi_result

# --- CONNECTING IT UP (DEMONSTRATION) ---
# We use widgets.interactive_output to automatically run a function
# whenever the 'year_slicer_widget' changes.

def update_dashboard_demo(selected_years):
    # This function automatically receives the new variable whenever you click the slicer.
    # You can pass this variable to any number of KPI functions here.

    # Example using the Economy data (assuming it exists)
    if 'Economy' in dataframes:
        calculate_example_kpi(dataframes['Economy'], selected_years)
    else:
        print(f"Selected variables: {selected_years}. (Economy data not found for demo).")

# Link the widget to the demo function
out = widgets.interactive_output(update_dashboard_demo, {'selected_years': year_slicer_widget})

# Display the output area (go back up and change the slicer selection to see this update!)
display(out)
```

```python
# @title
# Page 02 - Adult literacy rate % - function embedded in single function below also
# --- KPI DEFINITION: National Adult Literacy Rate ---
def calculate_national_adult_literacy(df_education, selected_years_tuple):
    """
    Calculates the Average National Adult Literacy Rate for the selected years.
    Logic: Average of 'Adult Literacy Rate - Total' where Region is 'OVERALL'.
    """

    # 1. Create a working copy of the dataframe
    df = df_education.copy()

    # 2. Rename the specific column we need (Unnamed: 13 -> Adult_Lit_Total)
    # Based on your verification: 'Unnamed: 13' is the Total (T) column
    df.rename(columns={'Unnamed: 13': 'Adult_Lit_Total'}, inplace=True)

    # 3. Filter for National Level data (Region = "OVERALL")
    # We strip whitespace to be safe (' OVERALL ' -> 'OVERALL')
    df_national = df[df['Region'].astype(str).str.strip() == 'OVERALL'].copy()

    # 4. Clean the Data: Convert column to numeric
    # 'errors=coerce' turns text headers (like 'T') into NaN so they don't crash the math
    df_national['Adult_Lit_Total'] = pd.to_numeric(df_national['Adult_Lit_Total'], errors='coerce')

    # 5. Apply the Dynamic Year Filter
    # We use the helper function you created in the previous step
    df_filtered = filter_by_selected_years(df_national, selected_years_tuple, year_col_name='Year')

    # 6. Calculate the KPI
    # Power BI Logic: AVERAGE(Education[Adult Literacy Rate - T])
    # The mean() function ignores NaNs automatically
    kpi_value = df_filtered['Adult_Lit_Total'].mean()

    # 7. Display the Result
    # If no data is found for the selected years, handle it gracefully
    if pd.isna(kpi_value):
        print(f"National Adult Literacy Rate: No Data for {selected_years_tuple}")
    else:
        print(f"National Adult Literacy Rate: {kpi_value:.2f}%")

    return kpi_value

# --- LINKING TO SLICER ---
# This updates the 'update_dashboard_demo' function to include this new KPI.
def update_dashboard(selected_years):
    print(f"--- DASHBOARD UPDATE (Selected Range: {selected_years}) ---")

    # Check if file is loaded
    if 'Education' in dataframes:
        calculate_national_adult_literacy(dataframes['Education'], selected_years)
    else:
        print("Error: 'Education.xlsx' not loaded.")

# Re-display the interactive output with the new KPI
output_display = widgets.interactive_output(update_dashboard, {'selected_years': year_slicer_widget})
display(output_display)
```

```python
# @title
# Full Imunization % - Function embedded in single function below also
# --- KPI DEFINITION: National Full Immunization ---
def calculate_national_full_immunization(df_health, selected_years_tuple):
    """
    Calculates the Average National Full Immunization Rate for the selected years.
    Logic: Average of 'FULL IMMUNISATION - Total' where Region is 'OVERALL'.
    """

    # 1. Create a working copy
    df = df_health.copy()

    # 2. Rename the specific column we identified
    # 'Unnamed: 10' is the Total (T) column for Full Immunization
    df.rename(columns={'Unnamed: 10': 'Full_Imm_Total'}, inplace=True)

    # 3. Filter for National Level data (Region = "OVERALL")
    df_national = df[df['Region'].astype(str).str.strip() == 'OVERALL'].copy()

    # 4. Clean the Data: Convert column to numeric
    # This removes any headers or text values that might cause errors
    df_national['Full_Imm_Total'] = pd.to_numeric(df_national['Full_Imm_Total'], errors='coerce')

    # 5. Apply the Dynamic Year Filter
    # Uses the master slicer variable to filter rows
    df_filtered = filter_by_selected_years(df_national, selected_years_tuple, year_col_name='Year')

    # 6. Calculate the KPI
    # Corresponds to: AVERAGE(Health[FULL IMMUNISATION - T])
    kpi_value = df_filtered['Full_Imm_Total'].mean()

    # 7. Display the Result
    if pd.isna(kpi_value):
        print(f"National Full Immunization: No Data for {selected_years_tuple}")
    else:
        print(f"National Full Immunization: {kpi_value:.2f}%")

    return kpi_value

# --- LINKING TO DASHBOARD ---
def update_dashboard(selected_years):
    print(f"--- DASHBOARD UPDATE (Selected Range: {selected_years}) ---")

    # 1. Calculate Literacy KPI (from previous step)
    if 'Education' in dataframes:
        calculate_national_adult_literacy(dataframes['Education'], selected_years)

    # 2. Calculate Immunization KPI (New)
    if 'Health' in dataframes:
        calculate_national_full_immunization(dataframes['Health'], selected_years)
    else:
        print("Error: 'Health.xlsx' not loaded.")

# Re-display the master interactive block
output_display = widgets.interactive_output(update_dashboard, {'selected_years': year_slicer_widget})
display(output_display)
```

```python
# @title
# Adult literacy, immunization and home owners function - embedded below also
# --- KPI DEFINITION: % Home Owners ---
def calculate_home_ownership(df_household, selected_years_tuple):
    """
    Calculates the Average % Home Ownership for the selected years.
    Logic: Average of 'OWN' column in the Tenure section where Region is 'OVERALL'.
    """

    # 1. Extract only the relevant columns for Housing Tenure (Indices 12, 13, 14)
    # Index 12 = Year.2
    # Index 13 = Region
    # Index 14 = DISTRIBUTIONS... (which contains 'OWN' data)
    df_tenure = df_household.iloc[:, [12, 13, 14]].copy()

    # 2. Rename columns for clarity
    df_tenure.columns = ['Year', 'Region', 'Own_Percentage']

    # 3. Drop the first row (Row 0) which contains the sub-header "OWN"
    df_tenure = df_tenure.iloc[1:].copy()

    # 4. Filter for National Level data (Region = "OVERALL")
    df_national = df_tenure[df_tenure['Region'].astype(str).str.strip() == 'OVERALL'].copy()

    # 5. Clean the Data: Convert percentage column to numeric
    df_national['Own_Percentage'] = pd.to_numeric(df_national['Own_Percentage'], errors='coerce')

    # 6. Apply the Dynamic Year Filter
    # Note: We use the 'Year' column we renamed in step 2
    df_filtered = filter_by_selected_years(df_national, selected_years_tuple, year_col_name='Year')

    # 7. Calculate the KPI
    # Corresponds to: AVERAGE(Household_Ownership[Percentage]) where Type="OWN"
    kpi_value = df_filtered['Own_Percentage'].mean()

    # 8. Display the Result
    if pd.isna(kpi_value):
        print(f"Home Owners: No Data for {selected_years_tuple}")
    else:
        print(f"Home Owners: {kpi_value:.2f}%")

    return kpi_value

# --- LINKING TO DASHBOARD ---
def update_dashboard(selected_years):
    print(f"--- DASHBOARD UPDATE (Selected Range: {selected_years}) ---")

    # 1. Literacy KPI
    if 'Education' in dataframes:
        calculate_national_adult_literacy(dataframes['Education'], selected_years)

    # 2. Immunization KPI
    if 'Health' in dataframes:
        calculate_national_full_immunization(dataframes['Health'], selected_years)

    # 3. Home Ownership KPI (New)
    if 'Household' in dataframes:
        calculate_home_ownership(dataframes['Household'], selected_years)
    else:
        print("Error: 'Household.xlsx' not loaded.")

# Re-display the master interactive block
output_display = widgets.interactive_output(update_dashboard, {'selected_years': year_slicer_widget})
display(output_display)
```

```python
# # Adult literacy, immunization home owners and avg national income function

# --- KPI DEFINITION: Average National Income ---
def calculate_avg_national_income(df_economy, selected_years_tuple):
    """
    Calculates the Average National Income for the selected years.
    Logic: Average of 'Avg Income per HouseHold' where Region is 'Pakistan'.
    """

    # 1. Create a working copy
    df = df_economy.copy()

    # 2. Filter for National Level data (Region = "Pakistan")
    # We strip whitespace to be safe ('Pakistan ' -> 'Pakistan')
    df_national = df[df['Region'].astype(str).str.strip() == 'Pakistan'].copy()

    # 3. Clean the Data: Convert Income column to numeric
    # The column name is exactly 'Avg Income per HouseHold' as per your inspection
    col_name = 'Avg Income per HouseHold'
    df_national[col_name] = pd.to_numeric(df_national[col_name], errors='coerce')

    # 4. Apply the Dynamic Year Filter
    df_filtered = filter_by_selected_years(df_national, selected_years_tuple, year_col_name='Year')

    # 5. Calculate the KPI
    # Corresponds to: AVERAGE(Economy[Avg Income per HouseHold])
    kpi_value = df_filtered[col_name].mean()

    # 6. Display the Result
    if pd.isna(kpi_value):
        print(f"Avg National Income: No Data for {selected_years_tuple}")
    else:
        # We format with commas for currency (e.g., 7,167.00)
        print(f"Avg National Income: {kpi_value:,.2f}")

    return kpi_value

# --- LINKING TO DASHBOARD ---
def update_dashboard(selected_years):
    print(f"--- DASHBOARD UPDATE (Selected Range: {selected_years}) ---")

    # 1. Literacy KPI
    if 'Education' in dataframes:
        calculate_national_adult_literacy(dataframes['Education'], selected_years)

    # 2. Immunization KPI
    if 'Health' in dataframes:
        calculate_national_full_immunization(dataframes['Health'], selected_years)

    # 3. Home Ownership KPI
    if 'Household' in dataframes:
        calculate_home_ownership(dataframes['Household'], selected_years)

    # 4. Avg National Income KPI (New)
    if 'Economy' in dataframes:
        calculate_avg_national_income(dataframes['Economy'], selected_years)
    else:
        print("Error: 'Economy.xlsx' not loaded.")

# Re-display the master interactive block
output_display = widgets.interactive_output(update_dashboard, {'selected_years': year_slicer_widget})
display(output_display)
```

```python
#urban vs rural basic access disparity check

import plotly.graph_objects as go
import pandas as pd
import numpy as np

# --- 1. GET SELECTION ---
current_selection = year_slicer_widget.value
if not current_selection:
    # Default to all years if nothing selected
    df_temp = dataframes['Household']
    current_selection = tuple(df_temp['Year'].dropna().unique())

print(f"Generating Disparity Chart for: {current_selection}")

# --- 2. SETUP ---
df = dataframes['Household'].copy()
categories = ['Electricity', 'Flush Toilet', 'Tap Water']
urban_vals = []
rural_vals = []

# --- 3. DATA EXTRACTION ---

# [A] ELECTRICITY (Cols 20-22)
# Structure: Year(20) | Region(21) | Value(22)
df_elec = df.iloc[:, [20, 21, 22]].copy()
df_elec.columns = ['Year', 'Region', 'Percentage']
df_elec.loc[:, 'Year'] = df_elec['Year'].ffill() # Fix Warning

# Filter
subset_elec = df_elec[df_elec['Year'].isin(current_selection)].copy()
subset_elec.loc[:, 'Percentage'] = pd.to_numeric(subset_elec['Percentage'], errors='coerce')

# Extract Values (Use MAX to be safe)
u_elec = subset_elec[subset_elec['Region'].astype(str).str.strip().str.lower() == 'urban']['Percentage'].max()
r_elec = subset_elec[subset_elec['Region'].astype(str).str.strip().str.lower() == 'rural']['Percentage'].max()

urban_vals.append(u_elec)
rural_vals.append(r_elec)


# [B] FLUSH TOILET (Cols 6-9)
# Structure: Year(6) | Source(7) | Urban(8) | Rural(9)
# REMOVED: The check for 'MainRegion' (Col 1) as it was incorrect.
df_toilet = df.iloc[:, [6, 7, 8, 9]].copy()
df_toilet.columns = ['Year', 'Source', 'Urban', 'Rural']
df_toilet.loc[:, 'Year'] = df_toilet['Year'].ffill()

# Filter: Year + Source="FLUSH"
subset_toilet = df_toilet[
    (df_toilet['Year'].isin(current_selection)) &
    (df_toilet['Source'].astype(str).str.strip() == 'FLUSH')
].copy()

subset_toilet.loc[:, 'Urban'] = pd.to_numeric(subset_toilet['Urban'], errors='coerce')
subset_toilet.loc[:, 'Rural'] = pd.to_numeric(subset_toilet['Rural'], errors='coerce')

urban_vals.append(subset_toilet['Urban'].max())
rural_vals.append(subset_toilet['Rural'].max())


# [C] TAP WATER (Cols 0-3)
# Structure: Year(0) | Source(1) | Urban(2) | Rural(3)
df_water = df.iloc[:, [0, 1, 2, 3]].copy()
df_water.columns = ['Year', 'Source', 'Urban', 'Rural']
df_water.loc[:, 'Year'] = df_water['Year'].ffill()

# Filter: Year + Source="PIPED WATER"
subset_water = df_water[
    (df_water['Year'].isin(current_selection)) &
    (df_water['Source'].astype(str).str.contains('PIPED', case=False, na=False))
].copy()

subset_water.loc[:, 'Urban'] = pd.to_numeric(subset_water['Urban'], errors='coerce')
subset_water.loc[:, 'Rural'] = pd.to_numeric(subset_water['Rural'], errors='coerce')

urban_vals.append(subset_water['Urban'].max())
rural_vals.append(subset_water['Rural'].max())

# --- 4. CLEANUP & FEEDBACK ---
urban_vals = [0 if pd.isna(x) else x for x in urban_vals]
rural_vals = [0 if pd.isna(x) else x for x in rural_vals]

print("\n--- Final Data Check ---")
print(f"Electricity: Urban={urban_vals[0]:.1f}%, Rural={rural_vals[0]:.1f}%")
print(f"Flush Toilet: Urban={urban_vals[1]:.1f}%, Rural={rural_vals[1]:.1f}%")
print(f"Tap Water:   Urban={urban_vals[2]:.1f}%, Rural={rural_vals[2]:.1f}%")

# --- 5. PLOT CHART ---
fig = go.Figure()

fig.add_trace(go.Bar(
    x=categories,
    y=urban_vals,
    name='Urban',
    marker_color='#114f1e',
    text=[f'{x:.0f}%' if x > 0 else '' for x in urban_vals],
    textposition='auto'
))

fig.add_trace(go.Bar(
    x=categories,
    y=rural_vals,
    name='Rural',
    marker_color='#8bc34a',
    text=[f'{x:.0f}%' if x > 0 else '' for x in rural_vals],
    textposition='auto'
))

fig.update_layout(
    title='<b>Urban vs. Rural Basic Access Disparity</b>',
    xaxis_title='',
    yaxis_title='Percentage (%)',
    barmode='group',
    plot_bgcolor='white',
    legend=dict(orientation="h", y=1.02, x=1),
    height=500,
    yaxis=dict(range=[0, 105])
)

fig.show()
```

```python
#Vulnerable employment share and comparison with Pakistan overall - Average

import pandas as pd
import plotly.graph_objects as go

# ============================================================
# 1. GET YEAR SELECTION
# ============================================================
current_selection = year_slicer_widget.value

if not current_selection:
    if 'Employment' in dataframes:
        df_temp = dataframes['Employment']
        current_selection = tuple(df_temp.iloc[:, 0].dropna().unique())


# ============================================================
# 2. LOAD & PREPARE DATA
# ============================================================
if 'Employment' not in dataframes:
    print("Error: Employment data missing.")
else:
    df = dataframes['Employment'].copy()

    df_emp = df.iloc[:, [0, 1, 4, 5, 6]].copy()
    df_emp.columns = [
        'Year', 'Area',
        'Agriculture', 'NonAgriculture', 'Unpaid'
    ]

    # Clean Year and numeric data
    df_emp['Year'] = df_emp['Year'].ffill()
    for col in ['Agriculture', 'NonAgriculture', 'Unpaid']:
        df_emp[col] = (
            df_emp[col].astype(str)
            .str.replace(',', '', regex=False)
        )
        df_emp[col] = pd.to_numeric(df_emp[col], errors='coerce').fillna(0)

    # Business logic
    df_emp['Vulnerable Employment'] = (
        df_emp['Agriculture'] + df_emp['Unpaid']
    )
    df_emp['Formal Employment'] = df_emp['NonAgriculture']

    # ============================================================
    # 3. AGGREGATION
    # ============================================================

    # Urban & Rural (bars)
    df_ur = df_emp[
        (df_emp['Area'].isin(['Urban', 'Rural'])) &
        (df_emp['Year'].isin(current_selection))
    ]

    df_ur_avg = (
        df_ur.groupby('Area')[['Formal Employment', 'Vulnerable Employment']]
        .mean()
        .reset_index()
    )

    # Pakistan benchmark (single national average)
    df_pk = df_emp[
        (df_emp['Area'] == 'Pakistan') &
        (df_emp['Year'].isin(current_selection))
    ]

    pk_formal = df_pk['Formal Employment'].mean()
    pk_vulnerable = df_pk['Vulnerable Employment'].mean()

    # ============================================================
    # 4. PLOT
    # ============================================================

    fig = go.Figure()

    # Formal (bars)
    fig.add_bar(
        x=df_ur_avg['Area'],
        y=df_ur_avg['Formal Employment'],
        name='Formal Employment',
        marker_color='#0D47A1',
        text=[f'{v:.1f}%' for v in df_ur_avg['Formal Employment']],
        textposition='outside'
    )

    # Vulnerable (bars)
    fig.add_bar(
        x=df_ur_avg['Area'],
        y=df_ur_avg['Vulnerable Employment'],
        name='Vulnerable Employment',
        marker_color='#E65100',
        text=[f'{v:.1f}%' for v in df_ur_avg['Vulnerable Employment']],
        textposition='outside'
    )

    # Pakistan benchmark lines
    fig.add_hline(
        y=pk_formal,
        line_dash='dash',
        line_color='#0D47A1',
        annotation_text=f'Pakistan Formal ({pk_formal:.1f}%)',
        annotation_position='top left'
    )

    fig.add_hline(
        y=pk_vulnerable,
        line_dash='dash',
        line_color='#E65100',
        annotation_text=f'Pakistan Vulnerable ({pk_vulnerable:.1f}%)',
        annotation_position='bottom left'
    )

    ymax = max(
        df_ur_avg[['Formal Employment', 'Vulnerable Employment']].max().max(),
        pk_formal,
        pk_vulnerable
    )

    fig.update_layout(
        title='<b>Employment Structure: Urban & Rural vs Pakistan Benchmark</b>',
        xaxis_title='Region',
        yaxis_title='Average Share (%)',
        barmode='group',
        plot_bgcolor='white',
        yaxis=dict(
            gridcolor='lightgrey',
            range=[0, ymax * 1.3]
        ),
        legend=dict(
            orientation='h',
            y=1.12,
            x=0.5,
            xanchor='center'
        ),
        height=550
    )

    fig.show()
```

```python
#Household access to core infrastructure

import plotly.graph_objects as go
import pandas as pd
import numpy as np

# --- 1. ROBUST STATISTICAL AVERAGE FUNCTION ---
def get_verified_avg(df, year_col_idx, label_col_idx, val_col_idx, target_label, selected_years):
    # Extract specific column slice
    temp_df = df.iloc[:, [year_col_idx, label_col_idx, val_col_idx]].copy()
    temp_df.columns = ['Year', 'Label', 'Value']

    # 1. Forward fill Years to fix merged Excel cells
    temp_df['Year'] = temp_df['Year'].ffill()

    # 2. Convert to numeric, turning strings/blanks into NaN
    temp_df['Value'] = pd.to_numeric(temp_df['Value'], errors='coerce')

    # 3. Apply Filters: Must match 'TOTAL' or specific category exactly
    # We use .str.contains to handle potential trailing spaces in Excel
    mask = (temp_df['Year'].isin(selected_years)) & \
           (temp_df['Label'].astype(str).str.contains(target_label, case=False, na=False))

    # 4. Statistical Mean: .mean() in pandas automatically ignores NaNs (matches Power BI)
    return temp_df[mask]['Value'].mean()

# --- 2. EXECUTION ---
if 'Household' in dataframes:
    df_raw = dataframes['Household']
    # Use your selected years from the slicer
    years = year_slicer_widget.value or ['2004-2005']

    # Calculate each pillar using the verified offsets from your diagnostic check
    # Electricity: Index 21 is Region, 22 is %
    elec_acc = get_verified_avg(df_raw, 20, 21, 22, 'TOTAL', years)

    # Flush Toilet: Index 7 is Source, 10 is %
    toilet_acc = get_verified_avg(df_raw, 6, 7, 10, 'FLUSH', years)

    # Tap Water: Index 1 is Source, 3 is % (Labels as 'PIPED WATER')
    water_acc = get_verified_avg(df_raw, 0, 1, 3, 'PIPED WATER', years)

    # --- 3. PLOT DONUT ---
    vals = [elec_acc, toilet_acc, water_acc]
    labels = ['Electricity', 'Flush Toilet', 'Tap Water']

    fig = go.Figure(data=[go.Pie(
        labels=labels, values=vals, hole=.5,
        marker=dict(colors=['#FFD600', '#1A237E', '#03A9F4']),
        textinfo='label+percent'
    )])

    fig.update_layout(
        title=f'<b>Core Infrastructure Access Portfolio</b><br>Average: {", ".join(years)}',
        annotations=[dict(text='Access', x=0.5, y=0.5, font_size=20, showarrow=False)],
        height=550,
        legend=dict(orientation="h", y=-0.1, x=0.5, xanchor="center")
    )
    fig.show()

    print(f"Verified Stats: Electricity: {elec_acc:.1f}% | Toilet: {toilet_acc:.1f}% | Water: {water_acc:.1f}%")
```
